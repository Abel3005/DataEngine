### 라우팅 구조
기본적으로 url의 트리 구조를 따라서 요청된 URL주소 및 데이터를 View에 전송하고 해당되는 Template 화면을 출력
- Template 문법을 따라서 View의 동작을 명식적으로 호출 할 수 있습니다. 
>   - 애플리케이션과 뷰의 이름을 식별할 수 있도록 해야 합니다.
>   > - 연결된 애플리케이션의 url에서 app_name 변수를 정의해줘야 합니다.  
>   > - urlpatterns에서 View의 로직과 연결할 때, 해당 연결에 대한 이름을 지정해 줘야 합니다.
>   > - 해당 View의 동작에 필요한 데이터를 인자로서 주어야 합니다.  
>   - ```{% url "[app_name]:[name]" 데이터...  %}```  
>   > - 예를 들어 ```{% url "polls:detail" question.id  %}```와 같이 polls:detail로 할당된 View의 기능에 quesion.id 라는 인자를 주어야 합니다.
>   > - 무사히 실행시키기 위해서는 url이 구분되는 것을 필요로 합니다.

**알아두어야 할 사항**  
Django는 reverse 함수를 사용하여, url과 함께 몇몇의 데이터 인자들을 포함하여 URL 패턴을 만듭니다.
그러므로 HTTP 프로토콜로 리다이렉션 하는 연산을 할 때에 따로 args 인자를 주는 것이 아닌 다음과 같이 사용합니다.
>   - HttpResponseRedirect(reverse('polls:results', args=[]))
