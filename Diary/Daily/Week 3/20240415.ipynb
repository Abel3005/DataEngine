```python
def fibo(n):
    if n < 2:
        return n
    a,b = 0,1
    for i in range(n):
        a, b = b, a+b
    return a
```

```python
fibo(100)
```

```python
def fibo_recu(n):
    if n < 2:
        return n
    return fibo_recu(n-2) + fibo(n-1)
```

```python
def fibo_generator():
    a,b = 0,2
    while True:
        yield b
        a,b = b, a+b
```
